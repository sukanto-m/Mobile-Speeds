## Top Download Speeds by Service Provider

```python
df2 = df.loc[df["Download_Upload"] == "download"]
df2.head(10)
```

```python
df2.dropna(axis = 'rows')
```

```python
df3 = df2.sort_values("Data_Speed.Kbps.", axis = 0, ascending = False)
df3top = df3.head(10)
df3top
```

```python
df3top.mean(axis = 0)
```

```python
df3top.plot.barh(x = "Service_Area", y = "Data_Speed.Kbps.", title = "Top Download Speeds by Service Area")
plt.savefig('topdownload.png')
```

## Top Upload Speeds by Service Provider

```python
dfup = df.loc[df["Download_Upload"] == "upload"]
dfup.head(10)
```

```python
dfupclean= dfup.dropna(axis = 'rows')
dfupclean.head(10)
```

```python
dfupclean_top = dfupclean.sort_values("Data_Speed.Kbps.", axis = 0, ascending = False)
dfupclean_top.head(10)
```

```python
dfupclean_top = dfupclean_top.drop_duplicates(subset = 'Service_Area', keep = 'first', inplace = False)
dfupclean_top.head(10)
```

```python
dfupclean_top.mean(axis = 0)
```

```python
dfupclean_top.plot.barh(x = "Service_Area", y = "Data_Speed.Kbps.", title = "Top Upload Speeds by Service Area")
plt.savefig('topupload.png')
```

```python
dfupclean_top.to_excel('upload.xlsx')
```
